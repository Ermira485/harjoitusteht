```python
number1 = 5
number2 = 10
```

```python
if number1 == number2:
    print("number1 and number2 are equal")
```

```python
if number1 > number2:
    print("number1 is greater than number2")
```

```python
if number1 >= number2:
    print("number1 is greater than or equal to number2")
```

```python
if number1 != number2:
    print("number1 and number2 are NOT equal")
```

```python
number1 = 5
number2 = 5
number3 = 5
```

```python
if number1 == number2 and number2 == number3:
    print("All three numbers are equal")
```

```python
if number1 == number2 or number2 == number3:
    print("Either number1==number2 or number2==number3 (or both)")
```

```python
if number1 > number2 and number1 > number3:
    print("number1 is greater than number2 AND number3")
```

```python
if number1 > number2:
    print("number1 is greater than number2")
else:
    if number2 > number3:
        print("number2 is greater than number3 (and number1 was not > number2)")
```

```python
if number1 == number2:
    print("number1 and number2 are equal")
else:
    if number1 == number3:
        print("number1 and number3 are equal (but number1 and number2 are not)")
```
